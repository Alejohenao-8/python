```
#Tu objetivo es limpiar los datos para que estén listos para el análisis, aplicando las siguientes técnicas:
import pandas as pd

data = {
    'Nombre': ['Juan', 'Maria', 'Pedro', 'Ana', 'Luisa', 'Carlos', 'Juan'],
    'Edad': [25, 30, None, 29, 28, 200, 25],
    'Email': ['juan@mail.com', 'maria@mail.com', 'pedro@mail.com', 'ana@mail.com', 'luisa@mail.com', 'carlos@mail.com', 'juan@mail.com'],
    'Pais': ['colombia', 'Colombia', 'Mexico', 'Argentina', 'Colombia', 'Brazil', 'colombia'],
    'Genero': ['M', 'F', 'H', 'F', 'Femenino', 'M', 'M'],
    'Visitas': [5, 7, 6, 4, 8, 2, 5]
}

df = pd.DataFrame(data)

print("Datos iniciales:")
print(df)
print("\n")

print("Columna 'Edad' después de rellenar valores faltantes:")
# Paso 3: Rellenar valores faltantes
df['Edad'].fillna(df['Edad'].mean(), inplace=True)

# Imprimimos la columna Edad para verificar que ya no hay valores faltantes
print(df)
print("\n")
# Paso 4: Eliminar duplicados
df.drop_duplicates(inplace=True)
print("\nEliminar duplicados")
print(df)
print("\n")

# Paso 5: Estandarizar los valores de las columnas
# Estandarizar el género
df['Genero'] = df['Genero'].replace({
    'M': 'Masculino',
    'H': 'Masculino',
    'F': 'Femenino',
    'Femenino': 'Femenino'
})

print("\nEstandarizar los valores de las columnas: género")
print(df)
print("\n")

# Estandarizar el país (asegurar que esté en mayúsculas)
df['Pais'] = df['Pais'].str.capitalize()

print("\nEstandarizar el país:")
print(df)
print("\n")

# Paso 6: Corregir valores inválidos (como la edad de 200 años).
# Asumimos que la edad máxima razonable es 100 años
df['Edad'] = df['Edad'].apply(lambda x: x if x < 100 else df['Edad'].mean())
print("\nCorregir valores inválidos:(como la edad de 200 años).")
print(df)
```

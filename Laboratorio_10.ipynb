```
#Ejercicio de práctica 1.
#1. Salud Digital y Análisis de Datos de Pacientes

import pandas as pd

# Crear el DataFrame inicial
data = {
    'Nombre': ['Carlos', 'Ana', 'Luis', 'Pedro', 'María', None],
    'Edad': [34, 29, None, 45, 25, 33],
    'Fecha_Ingreso': ['2021-01-15', '2020/03/12', '2022-07-01', '2021/12/01', '2021-05-20', '2020-03-12'],
    'Resultado_Prueba': ['Positivo', 'Negativo', 'Positivo', 'Positivo', 'Negativo', 'Negativo']
}

df = pd.DataFrame(data)
print("Datos de pacientes:")
print(df)
print("\n")

# 1. Rellenar valores faltantes en la columna 'Nombre' (si fuera necesario, se podría llenar con "Desconocido")
df['Nombre'].fillna('Desconocido', inplace=True)
print(" Rellenar valores faltantes en la columna 'Nombre' por Desconocido")
print(df)
print("\n")

# 2. Rellenar valores faltantes en la columna 'Edad' con la media de las edades
df['Edad'].fillna(df['Edad'].mean(), inplace=True)
print("valores faltantes en la columna 'Edad' con la media de las edades")
print(df)
print("\n")

# 3. Estandarizar el formato de la columna 'Fecha_Ingreso' a YYYY-MM-DD
df['Fecha_Ingreso'] = df['Fecha_Ingreso'].str.replace('/', '-')
print("Estandarizar el formato de la columna 'Fecha_Ingreso' a YYYY-MM-DD")
print(df)
print("\n")

```

```
#2. Comercio Electrónico y Recomendaciones Personalizadas

import pandas as pd

# Datos iniciales
data = {
    'Cliente_ID': [1, 2, 3, 4, 5],
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Cantidad': [10, 20, None, 5, 15],
    'Precio': ['100', '200$', '300', 'cuatrocientos', '500'],
    'Fecha_Transaccion': ['2023-01-15', '2023/02/20', '2023-03-25', '2023/04/30', '2023-05-10']
}

# Crear el DataFrame
df = pd.DataFrame(data)
print("Datos iniciales:")
print(df)
print("\n")

# Paso 1: Rellenar valores faltantes en la columna 'Cantidad'
# Rellenamos con la media de las cantidades
df['Cantidad'].fillna(df['Cantidad'].mean(), inplace=True)

print("valores faltantes en la columna 'Cantidad' con la media de las cantidades")
print(df)
print("\n")

# Paso 2: Limpiar y convertir la columna 'Precio' a valores numéricos
# Remover símbolos y textos y convertir a número, colocando NaN en caso de error
df['Precio'] = pd.to_numeric(df['Precio'].replace({'\$': '', 'cuatrocientos': '400'}, regex=True), errors='coerce')
print("Limpiar y convertir la columna 'Precio' a valores numéricos")
print(df)
print("\n")

# Paso 3: Estandarizar el formato de la columna 'Fecha_Transaccion' a YYYY-MM-DD
# Reemplazamos '/' por '-' y luego convertimos a formato datetime
df['Fecha_Transaccion'] = pd.to_datetime(df['Fecha_Transaccion'].str.replace('/', '-'), errors='coerce').dt.strftime('%Y-%m-%d')
print("Estandarizar el formato de la columna 'Fecha_Transaccion' a YYYY-MM-DD")
print(df)
print("\n")
```

```
#Escenario: eres un analista de datos en una pequeña empresa que está recopilando información sobre sus clientes.
#Tu objetivo es limpiar los datos para que estén listos para el análisis, aplicando las siguientes técnicas:
import pandas as pd

data = {
    'Nombre': ['Juan', 'Maria', 'Pedro', 'Ana', 'Luisa', 'Carlos', 'Juan'],
    'Edad': [25, 30, None, 29, 28, 200, 25],
    'Email': ['juan@mail.com', 'maria@mail.com', 'pedro@mail.com', 'ana@mail.com', 'luisa@mail.com', 'carlos@mail.com', 'juan@mail.com'],
    'Pais': ['colombia', 'Colombia', 'Mexico', 'Argentina', 'Colombia', 'Brazil', 'colombia'],
    'Genero': ['M', 'F', 'H', 'F', 'Femenino', 'M', 'M'],
    'Visitas': [5, 7, 6, 4, 8, 2, 5]
}

df = pd.DataFrame(data)

print("Datos iniciales:")
print(df)
print("\n")

print("Columna 'Edad' después de rellenar valores faltantes:")
# Paso 3: Rellenar valores faltantes
df['Edad'].fillna(df['Edad'].mean(), inplace=True)

# Imprimimos la columna Edad para verificar que ya no hay valores faltantes
print(df)
print("\n")
# Paso 4: Eliminar duplicados
df.drop_duplicates(inplace=True)
print("\nEliminar duplicados")
print(df)
print("\n")

# Paso 5: Estandarizar los valores de las columnas
# Estandarizar el género
df['Genero'] = df['Genero'].replace({
    'M': 'Masculino',
    'H': 'Masculino',
    'F': 'Femenino',
    'Femenino': 'Femenino'
})

print("\nEstandarizar los valores de las columnas: género")
print(df)
print("\n")

# Estandarizar el país (asegurar que esté en mayúsculas)
df['Pais'] = df['Pais'].str.capitalize()

print("\nEstandarizar el país:")
print(df)
print("\n")

# Paso 6: Corregir valores inválidos (como la edad de 200 años).
# Asumimos que la edad máxima razonable es 100 años
df['Edad'] = df['Edad'].apply(lambda x: x if x < 100 else df['Edad'].mean())
print("\nCorregir valores inválidos:(como la edad de 200 años).")
print(df)
```

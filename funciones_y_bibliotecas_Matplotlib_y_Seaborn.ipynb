```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Datos de ventas (ventas hipotticas)
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E', 'Producto F', 'Producto G', 'Producto H']
ciudades = ['Bogot', 'Cali', 'Bucaramanga']

# Crear un DataFrame con los datos de ventas para cada ciudad y producto
ventas = np.random.randint(100, 500, size=(8, 3))  # Generamos ventas aleatorias entre 100 y 500

# Crear un DataFrame de pandas para estructurar mejor los datos
df = pd.DataFrame(ventas, columns=ciudades, index=productos)

# Configuracin del grfico
sns.set(style="whitegrid")  # Establecer estilo atractivo
plt.figure(figsize=(10, 6))  # Ajustamos el tamao del grfico

# Crear el grfico de barras
df.plot(kind='bar', width=0.8, figsize=(10, 6))

# Aadir etiquetas y ttulo
plt.title('Ventas de productos por ciudad', fontsize=16)
plt.xlabel('Productos', fontsize=12)
plt.ylabel('Ventas', fontsize=12)

# Mostrar leyenda y el grfico
plt.legend(title='Ciudades', title_fontsize='13', loc='upper left')
plt.xticks(rotation=45, ha='right')

# Mostrar el grfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Datos de ventas (ventas hipotéticas)
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D', 'Producto E', 'Producto F', 'Producto G', 'Producto H']
ciudades = ['Bogotá', 'Cali', 'Bucaramanga']

# Crear un DataFrame con los datos de ventas para cada ciudad y producto
ventas = np.random.randint(100, 500, size=(8, 3))  # Generamos ventas aleatorias entre 100 y 500

# Crear un DataFrame de pandas para estructurar mejor los datos
df = pd.DataFrame(ventas, columns=ciudades, index=productos)

# Configuración del gráfico
sns.set(style="whitegrid")  # Establecer estilo atractivo
plt.figure(figsize=(10, 6))  # Ajustamos el tamaño del gráfico

# Crear el gráfico de barras
df.plot(kind='bar', width=0.8, figsize=(10, 6))

# Añadir etiquetas y título
plt.title('Ventas de productos por ciudad', fontsize=16)
plt.xlabel('Productos', fontsize=12)
plt.ylabel('Ventas', fontsize=12)

# Mostrar leyenda y el gráfico
plt.legend(title='Ciudades', title_fontsize='13', loc='upper left')
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Datos de ventas (ventas hipotéticas)
productos = ['Papa', 'Cebolla', 'Tomate']
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto']
ciudades = ['Bogotá', 'Cali', 'Bucaramanga']

# Generar ventas aleatorias para cada producto en cada ciudad a lo largo de los 8 meses
np.random.seed(42)  # Semilla para reproducibilidad
ventas = np.random.randint(50, 300, size=(3, 3, 8))  # Forma (3 productos, 3 ciudades, 8 meses)

# Crear una lista para organizar los datos en formato largo
data = []
for i, producto in enumerate(productos):
    for j, ciudad in enumerate(ciudades):
        for k, mes in enumerate(meses):
            data.append({
                'Producto': producto,
                'Ciudad': ciudad,
                'Mes': mes,
                'Ventas': ventas[i, j, k]
            })

# Convertir a un DataFrame de pandas
df = pd.DataFrame(data)

# Configuración del gráfico
sns.set(style="whitegrid")  # Establecer estilo atractivo
plt.figure(figsize=(12, 6))  # Ajustamos el tamaño del gráfico

# Crear el gráfico de líneas
sns.lineplot(x='Mes', y='Ventas', hue='Producto', style='Ciudad', markers=True, data=df, dashes=False)

# Añadir etiquetas y título
plt.title('Evolución de las Ventas de Productos a lo Largo de 8 Meses', fontsize=16)
plt.xlabel('Meses', fontsize=12)
plt.ylabel('Ventas', fontsize=12)

# Mostrar leyenda y ajustar la visualización
plt.legend(title='Productos', title_fontsize='13', loc='upper left')
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

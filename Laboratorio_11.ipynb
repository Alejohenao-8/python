```
#1. Comparación de calificaciones de estudiantes en diferentes asignaturas
#Se quiere comparar las calificaciones promedio de los estudiantes en distintas materias
#para detectar diferencias de rendimiento.

import matplotlib.pyplot as plt

# Definición de las asignaturas y las calificaciones
asignaturas = ['Matemáticas', 'Ciencias', 'Historia', 'Inglés', 'Arte']
calificaciones = [85, 90, 78, 88, 92]

# Crear un gráfico de barras para visualizar la comparación
plt.figure(figsize=(7, 5))
plt.bar(asignaturas, calificaciones, color='skyblue')

# Título y etiquetas
plt.title('Comparación de Calificaciones Promedio de Estudiantes', fontsize=14)
plt.xlabel('Asignaturas', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)

plt.show()
```

```
#2. Comparación del tiempo de carga de diferentes páginas web
#Se quiere visualizar el tiempo promedio de carga de diferentes sitios web para
#identificar cuál es más rápido o lento.
import matplotlib.pyplot as plt

# Definición de los sitios web y los tiempos de carga
sitios = ['Sitio A', 'Sitio B', 'Sitio C', 'Sitio D', 'Sitio E']
tiempos = [1.2, 2.5, 0.9, 3.0, 1.8]

# Crear un gráfico de barras para visualizar los tiempos de carga
plt.figure(figsize=(7, 5))
plt.bar(sitios, tiempos, color='lightgreen')

# Título y etiquetas
plt.title('Comparación de Tiempos de Carga de Páginas Web', fontsize=14)
plt.xlabel('Sitios Web', fontsize=12)
plt.ylabel('Tiempo de Carga (segundos)', fontsize=12)

# Mostrar los tiempos de carga encima de cada barra para mayor claridad
for i, tiempo in enumerate(tiempos):
    plt.text(i, tiempo + 0.1, f'{tiempo}s', ha='center', fontsize=10)

# Mostrar la gráfica
plt.show()
```

```
#3. Relación entre el número de horas de estudio y el rendimiento académico
#Se quiere ver si existe una correlación entre las horas de estudio y el rendimiento en los
#exámenes.
import matplotlib.pyplot as plt
import numpy as np

# Datos de horas de estudio y rendimiento en exámenes
horas_estudio = [2, 4, 6, 8, 10, 12, 14]
rendimiento = [55, 60, 65, 70, 75, 80, 85]

# Crear un gráfico de dispersión para visualizar la relación
plt.figure(figsize=(8, 6))
plt.scatter(horas_estudio, rendimiento, color='blue', label='Datos')

# Añadir la línea de mejor ajuste
m, b = np.polyfit(horas_estudio, rendimiento, 1)  # Calcular la pendiente (m) y la intersección (b)
plt.plot(horas_estudio, np.polyval([m, b], horas_estudio), color='red', label='Línea de mejor ajuste')

# Título y etiquetas
plt.title('Relación entre Horas de Estudio y Rendimiento Académico', fontsize=14)
plt.xlabel('Horas de Estudio', fontsize=12)
plt.ylabel('Rendimiento en Exámenes (%)', fontsize=12)

# Mostrar el coeficiente de correlación
correlacion = np.corrcoef(horas_estudio, rendimiento)[0, 1]
plt.text(10, 60, f'Correlación: {correlacion:.2f}', fontsize=12, color='black')

# Mostrar la gráfica
plt.legend()
plt.show()
```

```
#4. Distribución de salarios en tres departamentos distintos
#Se tiene un conjunto de datos de salarios en diferentes departamentos y quieres
#visualizar la variabilidad y mediana de los salarios en cada uno.
import matplotlib.pyplot as plt
import numpy as np

# Datos de salarios en tres departamentos
departamento_a = [3200, 3500, 3800, 4000, 3700, 3900, 4200]
departamento_b = [2200, 2400, 2500, 2800, 3000, 3200, 3300]
departamento_c = [5000, 5200, 5300, 5400, 5500, 5600, 5700]

# Crear un gráfico de caja (boxplot) para visualizar la distribución de los salarios
plt.figure(figsize=(5, 4))
plt.boxplot([departamento_a, departamento_b, departamento_c], labels=['Departamento A', 'Departamento B', 'Departamento C'])

# Título y etiquetas
plt.title('Distribución de Salarios en Tres Departamentos', fontsize=14)
plt.xlabel('Departamento', fontsize=12)
plt.ylabel('Salario', fontsize=12)

# Mostrar el gráfico
plt.show()

# Calcular la mediana y desviación estándar para cada departamento
mediana_a = np.median(departamento_a)
mediana_b = np.median(departamento_b)
mediana_c = np.median(departamento_c)

std_a = np.std(departamento_a)
std_b = np.std(departamento_b)
std_c = np.std(departamento_c)

# Imprimir los resultados
print(f"Departamento A - Mediana: {mediana_a}, Desviación Estándar: {std_a}")
print(f"Departamento B - Mediana: {mediana_b}, Desviación Estándar: {std_b}")
print(f"Departamento C - Mediana: {mediana_c}, Desviación Estándar: {std_c}")
```
